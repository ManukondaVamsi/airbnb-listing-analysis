### Objective 2. Prepare and reformat the data for visualization

#### i) create table = paris_listings_neighbourhood,
calcultes the mean price for each neighbourhood form lowest to highest
#### ii) create table = paris_listings_accomodations,
should be filtered down to the most expesive neighbourhood in paris,grouped by accomodations contains mean price 
#### iii) create table = paris_listing_over_time, 
is grouped by the year of the host_since column. count rows, total no. of hosts and average price of the year


```python
paris_listings_neighbourhood = (
    paris_listings
    .groupby("neighbourhood")
    .agg({"price": "mean"})
    .sort_values("price")
)
paris_listings_neighbourhood.tail()
```

```python
paris_listings_accommodates = (
    paris_listings
    .query("neighbourhood == 'Elysee'")
    .groupby("accommodates")
    .agg({"price": "mean", "neighbourhood":"count"})
    .sort_values("price")
)

paris_listings_accommodates.head()
```

```python
paris_listing_over_time = (
    paris_listings
    .set_index("host_since")
    .resample("YE")
    .agg({
        "neighbourhood":"count",
        "price":"mean"

    })
)
paris_listing_over_time.head()
```

```python
import seaborn as sns

(paris_listings_neighbourhood
 .plot
 .barh(
     title = "Average Lisitngs Price by Paris Neighbourhood",
     xlabel = "Price Per Night (EUR)",
     ylabel = "Neighbourhood",
     legend = None
 )

)
sns.despine()
```

```python
import seaborn as sns

(paris_listings_accommodates
 .plot
 .barh(
     title = "Average Lisitngs Price by Paris Acomodation Number",
     xlabel = "Price Per Night (EUR)",
     ylabel = "Acomodation capacity ",
     #legend = None
 )

)
sns.despine()
```

```python
paris_listing_over_time["neighbourhood"].plot(
    ylabel="New Hosts",
    title="New AirBnb Hosts in Paris Over time "
)
sns.despine()
```

```python
paris_listing_over_time["price"].plot(
    ylabel="Average price (EUR)",
    title="Average AirBnb price in Paris Over time "
)
sns.despine()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots()

ax.plot(
    paris_listing_over_time.index,
    paris_listing_over_time["neighbourhood"],  
    label="New Hosts",
    color="pink"
)
ax.set_ylabel("New Hosts")
ax2 = ax.twinx()

ax2.plot(
    paris_listing_over_time.index,
    paris_listing_over_time["price"],  
    label="Average Price"
)
ax2.set_ylim(0)
ax2.set_ylabel("Average Price")

ax2.set_title("2015 Regulations Lead to Few New Hosts,  Higher Prices")
```
